```
import requests
from bs4 import BeautifulSoup
import csv
from datetime import datetime
```

```
page = requests.get("https://www.chicagoreader.com/chicago/movie-reviews-long-form-film-criticism-index/Content?oid=5648188")
page
```

```
soup = BeautifulSoup(page.content, 'html.parser')
```

```
a = soup.select('p a')
```

```
len(a)
```

```
a
```

```
url=[]
for i in a:
  try:
    if '#top' in i['href']:
        a.pop('i')
        continue;
    if i['href'][:8]=='/chicago':
      url.append('https://www.chicagoreader.com'+i['href'])
    else:
      url.append(i['href'])
  except:
    pass
print(len(url))
```

```
url = url[5:-3]
```

```
len(url)
```

```
url
```

```
len(url)
```

```
with open('chicago_reader_review_urls_table_export.csv', mode='w') as csv_file:
    fieldnames = ['chicago_reader_href','date_created','date_updated','username','is_scraped']
    writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
    writer.writeheader()
    for i in url:
      writer.writerow({'chicago_reader_href': i, 'date_created': datetime.now(), 'date_updated': datetime.now(),'username':'Prathyusha'})
    
```
